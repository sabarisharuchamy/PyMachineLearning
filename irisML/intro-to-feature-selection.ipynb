```python
from sklearn import neighbors, datasets

#load the Iris dataset
iris = datasets.load_iris()

# iris
```

So that's the basic Iris dataset as it's stored in scikit-learn. It's a dictionary with a few keys that are interesting to us...

- `data` is an array of arrays (a 2-d array) containing the feature values for each of 150 examples.
- `feature_names` is an array of strings. These feature names correspond to the 'columns' in the 2-d `data` array.
- `target` is an array of classes, one for each 'row' of the `data` 2-d array

We're going to use [seaborn](https://seaborn.pydata.org/) to produce some neat statistical visualisations of the data. Seaborn works with [Pandas dataframes](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.html) so we'll need to create one from our raw Iris dictionary.

```python
# pandas is a lib for data analysis and modelling
import pandas as pd

# it uses "dataframes" - so prep a dataframe
iris_df = pd.DataFrame(data=iris.data[:,:], columns=iris.feature_names[:])

# add the classes to the dataframe as the last column
iris_df['target'] = pd.Series(iris.target)

# iris_df
```

That's a much neater representation of the Iris data. Pandas dataframes format themselves nicely for use in a notebook like this.

Now we'll put our first visualisation together - plotting two features against one another in a scatter plot.

```python
# seaborn is a statistical data visualisation package
import seaborn as sns

# it uses the venerable matplotlib
import matplotlib
import matplotlib.pyplot as plt

# set some visual stuff
sns.set(style="white", color_codes=True)

# plot the classes for two features at a time...
sns.FacetGrid(iris_df, hue="target", size=5) \
   .map(plt.scatter, "sepal length (cm)", "sepal width (cm)") \
   .add_legend()
```

We just plotted `sepal length (cm)` against `sepal width (cm)`.

Do you think a classifier is going to have an easier time classifying a data point that's in the blue class 0, or one that's in the red?

Note - classes 0, 1 and 2 map onto the three different kinds of Iris. In the data, they're labelled as numbers.

Try plotting other combinations of features (the others are `petal length (cm)` and `petal width (cm)`). Do you think some combinations of features are going to give us better classification accuracy than others?

"Overfitting" happens when a model learns details about a particular data set that don't describe the general case. When this happens, your accuracy will be better when you're testing than when you use your model for real. Compare  `sepal length (cm)` vs. `sepal width (cm)` with `petal length (cm)` vs. `petal width (cm)`. Which do you think is more likely to overfit if we're not careful?

We'll move on to actually using a k-nearest neighbour classifier to visualise the class boundaries that it learns.

```python
# set a value for k (no. of nearest neighbours)
k = 5

# create a new classifier instance
clf = neighbors.KNeighborsClassifier(k, weights='distance')

# see what defaults we got
# clf
```

We just created a new classifier, `clf`.

It doesn't work with dataframes and we'll be using numpy to generate a [meshgrid](https://docs.scipy.org/doc/numpy/reference/generated/numpy.meshgrid.html) of all points in the feature space.

```python
# visualise the areas that will "capture" a data point

# slice two features out of the data
feature_x_col = 'sepal length (cm)'
feature_y_col = 'petal length (cm)'

feature_x = iris_df.loc[:, feature_x_col]
feature_y = iris_df.loc[:, feature_y_col]

# and grab the target
target = iris_df.loc[:, 'target']

# feature_x
# feature_y
# target

# fit the model based on all the data for two features
clf.fit(iris_df.loc[:, [feature_x_col, feature_y_col]], target)
```

So we've got a classifier now that's been trained on all the data we have. Now we create a grid of every point at a given resolution (I chose 0.02 cm, you can update the `step` value if you want to) and get the classifier to classify every one.

```python
# make a grid so we can run the classifier over each point and colour it in with the predicted class
step = 0.02 # distance between points

# min and max values for the features
x_min, x_max = feature_x.min() - 1, feature_x.max() + 1
y_min, y_max = feature_y.min() - 1, feature_y.max() + 1

# generate a grid, using numpy
import numpy as np
xx, yy = np.meshgrid(np.arange(x_min, x_max, step),
                     np.arange(y_min, y_max, step))

# use the model to predict a class for all the points on the grid - flatten the 2d structure first
predictions = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# predictions

# now reshape the flattened structure back into the 2-d structure we need
predictions_xy = predictions.reshape(xx.shape)

# predictions_xy
```

OK, so what follows is basically a bunch of matplotlib commands to plot a 2-d scatter plot against the meshgrid, using specific colours (the `ListedColormap`s). You don't need to worry about the details, it's just using the matplotlib API rather than doing anything interesting.

```python
# plot the grid, predictions and training data

from matplotlib.colors import ListedColormap
cmap_light = ListedColormap(['#FFCCCC', '#CCFFCC','#CCCCFF'])
cmap_bold = ListedColormap(['#990000', '#009900','#000099'])

plt.figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
plt.pcolormesh(xx, yy, predictions_xy, cmap=cmap_light)
 
# Plot also the training points
plt.scatter(feature_x, feature_y, c=target, cmap=cmap_bold)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("3-Class classification (k = %i)" % (k))
plt.show()
```

So we can see the data points, along with more pastel colours that indicate which areas in the feature space would be classified as each of the three classes if there were a point there.

You can try the other features to see if the combinations you thought would perform well result in simple regions like the above, or more complex patterns.

It might be worth taking a moment to think about what this means. Remember, the labels and features here represent humans implementing a human classification system that experts have come up with, rather than laws of nature.

If you got this far, you're probably thinking "why not just run the classifier with every combination of feature possible and see how it does?" and you'd be right.

Let's do that!

```python
# now try all combinations of features, see which really works best!

from sklearn.model_selection import cross_val_score

from itertools import chain, combinations
i = set([0,1,2,3])
results = []
for feature_subset in chain.from_iterable(combinations(i, r) for r in range(len(i) + 1)):
    if feature_subset:
        med_score = np.median(cross_val_score(clf, iris.data[:, feature_subset], iris.target, cv=3));
        print('{:1.2f} {}'.format(med_score, feature_subset))
```

Interesting! A classifier using feature 3 alone achieves the best accuracy of any combination. A few other combinations can match it but do no better.

Certain features, when mixed with feature 3, actually reduce the accuracy of the classifier.

**Question** - I've done my feature selection on the whole data set, rather than only on my test data. What's wrong with that appoach?

**Answer** - The test data can influence the feature selection... but it's supposed to be "unseen". That can lead to overfitting. It doesn't matter for this example as we're just demonstrating some of the techniques and we don't really care about the final accuracy but it's something to be aware of for real world applications.
